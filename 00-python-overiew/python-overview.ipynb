### Branching

```python
a = int(input("Enter a number: "))
b = int(input("Enter another number: "))

res = a - b

if res > 0:
    print("The result was positive")
elif res < 0:
    print("The result is negative")
else:
    print("The result is zero")
```

### Looping

```python
for item in ["red", "green", "blue"]:
    print(item)
```

```python
for i in range(1, 11):
    print('5 X ', str(i), ' = ', str(5*i))
```

```python
i = 1
while i < 11:
    print('5 X ', str(i), ' = ', str(5*i))
    i += 1
```
